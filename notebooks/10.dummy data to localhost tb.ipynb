```python
import subprocess
# Define the command
mosquitto_pub_cmd = [
    "mosquitto_pub",
    "-d",           # Enable debugging output
    "-q", "1",       # Quality of Service level 1
    "-h", "localhost",
    "-p", "1883",
    "-t", "v1/devices/me/telemetry",
    "-u", "KSAeMr2upeZp4X4ikxEP",
    "-m", '{"temperature": 24}'  # Your telemetry data in JSON format
]

# Execute the command
try:
    subprocess.run(mosquitto_pub_cmd, check=True)
    print("Telemetry data published successfully.")
except subprocess.CalledProcessError as e:
    print(f"Error publishing telemetry data: {e}")
```

or use http from outside docker envirment

```python
import requests

# ThingsBoard API endpoint
api_endpoint = "http://localhost:8080/api/v1/KSAeMr2upeZp4X4ikxEP/telemetry"

# Telemetry data in JSON format
telemetry_data = {"temperature": 21}

# Set headers for the request
headers = {
    "Content-Type": "application/json",
}

# Perform the POST request
try:
    response = requests.post(api_endpoint, json=telemetry_data, headers=headers)
    response.raise_for_status()  # Raise an HTTPError for bad responses
    print("Telemetry data posted successfully.")
except requests.exceptions.RequestException as e:
    print(f"Error posting telemetry data: {e}")
    if 'response' in locals():
        print(f"Response content: {response.text}")
```
