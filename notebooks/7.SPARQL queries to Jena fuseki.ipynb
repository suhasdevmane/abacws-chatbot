ADDING SENSOR DATA TO APACHE JENA SERVER

```python
from rdflib import Graph, Literal, URIRef
from rdflib.namespace import RDF, RDFS, XSD
from SPARQLWrapper import SPARQLWrapper, JSON

# Connect to Fuseki server
from rdflib.plugins.stores import sparqlstore

# Replace 'http://localhost:3030/dataset' with your Fuseki endpoint
fuseki_endpoint = 'http://localhost:3030/dataset'
store = sparqlstore.SPARQLUpdateStore(queryEndpoint=fuseki_endpoint, update_endpoint=fuseki_endpoint + '/update')
graph = Graph(store)
```

```python
# Sensor information``
sensor_uri = URIRef('bldg:airq5.53')
room_uri = URIRef('bldg:Room5.53')
timestamp = "2023-11-03T12:00:00"  # Replace with your timestamp
sensor_reading = 25.5  # Replace with your sensor reading

# Build RDF triples
triple1 = (sensor_uri, RDF.type, Brick.Sensor)
triple2 = (sensor_uri, Brick.hasLocation, room_uri)
triple3 = (sensor_uri, Brick.isPointOf, room_uri)
triple4 = (sensor_uri, Brick.hasReading, Literal(sensor_reading, datatype=XSD.float))
triple5 = (sensor_uri, Brick.hasTimestamp, Literal(timestamp, datatype=XSD.dateTime))

# Add triples to the graph
graph.add(triple1)
graph.add(triple2)
graph.add(triple3)
graph.add(triple4)
graph.add(triple5)

# Commit the changes to the Fuseki server
graph.commit()

print(f"Sensor reading for {sensor_uri} added successfully at {timestamp}.")
```

```python
import json
from rdflib import Graph, URIRef, Literal, XSD

def convert_to_rdf(device_id, sensor_data):
    graph = Graph()

    for sensor, readings in sensor_data.items():
        for reading in readings:
            ts = reading["ts"]
            value = reading["value"]

            subject = URIRef(f"http://example.org/resource/{device_id}")
            predicate = URIRef(f"http://example.org/property/{sensor}")
            timestamp_literal = Literal(ts, datatype=XSD.dateTime)
            value_literal = Literal(value)

            graph.add((subject, predicate, value_literal, timestamp_literal))

    return graph

def send_to_fuseki(graph, fuseki_endpoint):
    graph.serialize(destination=fuseki_endpoint, format='xml')

# Example JSON data for a specific device
device_id = '478abf30-7db2-11ee-b0f3-69bd975277c1'
json_data = {
    "UV_light": [{"ts": 1700081712546, "value": "0.05"}],
    "Loudness": [{"ts": 1700081712546, "value": "348"}],
    "PM1.0Atmospheric": [{"ts": 1700081712546, "value": "10"}],
    # ... (add other sensors)
}

# Connect to Fuseki server
fuseki_endpoint = "http://localhost:3030/your_dataset/update"

# Convert JSON data to RDF and send to Fuseki
rdf_graph = convert_to_rdf(device_id, json_data)
send_to_fuseki(rdf_graph, fuseki_endpoint)
```
