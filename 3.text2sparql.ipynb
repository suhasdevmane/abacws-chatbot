```python

import json
import csv

# Read the JSON response from a file
with open('response.json', 'r', encoding='utf-8') as json_file:
    response_data = json.load(json_file)

# Extract data from the JSON response
bindings = response_data['results']['bindings']
headers = response_data['head']['vars']

# Extract data rows
data_rows = []
for binding in bindings:
    row = [binding.get(var, {}).get('value', '') for var in headers]
    data_rows.append(row)

# Specify the path to the CSV file
csv_file_path = 'output.csv'

# Write the data to the CSV file
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(headers)  # Write the header
    csv_writer.writerows(data_rows)  # Write the data rows

print(f'CSV file saved as: {csv_file_path}')


```
